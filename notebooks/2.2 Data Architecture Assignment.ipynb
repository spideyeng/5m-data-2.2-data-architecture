### Question 2 - Write Python code to list the first 10 objects (blob name) in the "gcp-public-data-landsat" bucket, along with their sizes.

```python
from google.cloud import storage

client = storage.Client()
bucket = client.get_bucket("gcp-public-data-landsat")
```

```python
print("Bucket name: {}".format(bucket.name))
print("Bucket location: {}".format(bucket.location))
print("Bucket storage class: {}".format(bucket.storage_class))
```

```python
list_blobs = bucket.list_blobs()

# Get first 10 blobs with their names and sizes
count = 0
for blob in list_blobs:
    if count >= 10:
        break
    print(f"Blob name: {blob.name}, Size: {blob.size} bytes")
    count += 1
```
